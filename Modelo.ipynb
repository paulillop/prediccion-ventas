```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv('datos_limpios.csv')
df.info()
```

# Convertir variables categóricas a numéricas.

## 1. Item_Identifier

```python
item_id = np.unique(df['Item_Identifier'])
gle = LabelEncoder()
itemid_labels = gle.fit_transform(df['Item_Identifier'])
itemid_mappings = {index: label for index, label in 
                  enumerate(gle.classes_)}
df['Item_Identifier'] =itemid_labels
df.head()
```

## 2. Item_Fat_Content

```python
df['Item_Fat_Content']=df['Item_Fat_Content'].replace(['Low Fat','Regular'],[1,0])
df.head()
```

## 3. Item_Type

```python
# transform and map
gen_le = LabelEncoder()
gen_labels = gen_le.fit_transform(df[['Item_Type']])
df['Item_Type_Label'] = gen_labels

# encode labels using one-hot encoding scheme
gen_ohe = OneHotEncoder()
gen_feature_arr = gen_ohe.fit_transform(
                              df[['Item_Type']]).toarray()
gen_feature_labels = list(gen_le.classes_)
gen_features = pd.DataFrame(gen_feature_arr, 
                            columns=gen_feature_labels)
gen_features
```

```python
df_1 = pd.concat([df, gen_features], axis=1)
df_1
```

## 4. Outlet_Identifier

```python
# transform and map
gen_le = LabelEncoder()
gen_labels = gen_le.fit_transform(df[['Outlet_Identifier']])
df_1['Outlet_Identifier_Label'] = gen_labels

# encode labels using one-hot encoding scheme
gen_ohe = OneHotEncoder()
gen_feature_arr = gen_ohe.fit_transform(
                              df[['Outlet_Identifier']]).toarray()
gen_feature_labels = list(gen_le.classes_)
gen_features = pd.DataFrame(gen_feature_arr, 
                            columns=gen_feature_labels)
df_1 = pd.concat([df_1, gen_features], axis=1)
df_1
```

## 5. Outlet_Size

```python
df_1['Outlet_Size'].value_counts()
```

```python
df_1['Outlet_Size']=df_1['Outlet_Size'].replace(['Small','Medium','High','S/I'],[1,2,3,0])
df_1
```

## 6. Outlet_Location_Type

```python
df_1['Outlet_Location_Type'].value_counts()
```

```python
df_1['Outlet_Location_Type']=df_1['Outlet_Location_Type'].replace(['Tier 1','Tier 2','Tier 3'],[1,2,3])
df_1.head()
```

## 7. Outlet_Type

```python
df_1['Outlet_Type'].value_counts()
```

```python
df_1['Outlet_Type']=df_1['Outlet_Type'].replace(['Supermarket Type1','Supermarket Type2','Supermarket Type3','Grocery Store'],[1,2,3,0])
```
